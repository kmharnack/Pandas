```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head(3)
```

```python
purchase_data_df.count()
```

## Player Count

* Display the total number of players


```python
#calculate unique players
unique_players = len(purchase_data_df["SN"].unique())
print(unique_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate values
unique_items = len(purchase_data_df["Item Name"].unique())
average_price = purchase_data_df["Price"].mean()
number_of_purchases = purchase_data_df["Purchase ID"].count()
total_revenue = purchase_data_df["Price"].sum()
```

```python
#create dataframe for unique items
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                          "Average Price": [average_price],
                          "Number of Purchases" : [number_of_purchases],
                          "Total Revenue" : [total_revenue]
})
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group results by gender
grouped_players_df=purchase_data_df.groupby(['Gender'])
grouped_players_df.count()
```

```python
#Create total unique counts for gender

#found nunique function here: https://stackoverflow.com/questions/38309729/count-unique-values-with-pandas-per-groups/38309823
#found to_frame function here: https://technology.amis.nl/2019/10/11/convert-groupby-result-on-pandas-data-frame-into-a-data-frame-using-to_frame/
total_count_df = grouped_players_df["SN"].nunique().to_frame(name = "Total Count")
total_count_df
```

```python
#create percentages and apply formatting
total_count_df["Percentage of Players"] = (total_count_df["Total Count"]/unique_players)
total_count_df["Percentage of Players"] = total_count_df["Percentage of Players"].astype(float).map("{:,.2%}".format)
total_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create analysis calculations and convert to dataframes
purchaseCount = purchase_data_df.groupby('Gender')['Purchase ID'].count()
purchaseCount_df = purchaseCount.to_frame(name = "Purchase Count")

totalValue = purchase_data_df.groupby('Gender')['Price'].sum()
totalValue_df = totalValue.to_frame(name = "Total Purchase Value")

app = totalValue/purchaseCount
app_df = app.to_frame(name = "Average Purchase Price")

avPerson = totalValue/(total_count_df['Total Count'])
avPerson_df = avPerson.to_frame(name = "Avg Total Purchase per Person")

gender_analysis_df = pd.concat([purchaseCount_df,app_df,totalValue_df,avPerson_df], axis = 1)
gender_analysis_df
```

```python
#apply formatting
gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
gender_analysis_df["Avg Total Purchase per Person"] = gender_analysis_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
gender_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create age bin
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,40]
bin_names = ["<10", "10-14", "15-19", "20-24","25-29", "30-34","35-39", "40+"]
purchase_data_df["Age Bins"] = pd.cut(purchase_data_df["Age"], bins, labels=bin_names, include_lowest=True)
age_bin_df = purchase_data_df.groupby(by = "Age Bins")
age_bin_df.count()
```

```python
#Slim to unique players and create dataframe
total_age_df = age_bin_df["SN"].nunique().to_frame(name = "Total Age Count")
total_age_df
```

```python
#create percentages and apply formatting
total_age_df["Percentage of Players"] = (total_age_df["Total Age Count"]/unique_players)
total_age_df["Percentage of Players"] =total_age_df["Percentage of Players"].astype(float).map("{:,.2%}".format)
total_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create analysis calculations and convert to dataframes
age_purchaseCount = purchase_data_df.groupby(by = "Age Bins")['Purchase ID'].count()
age_purchaseCount_df = age_purchaseCount.to_frame(name = "Purchase Count")

age_totalValue = purchase_data_df.groupby(by = "Age Bins")['Price'].sum()
age_totalValue_df = age_totalValue.to_frame(name = "Total Purchase Value")

age_app = age_totalValue/age_purchaseCount
age_app_df = age_app.to_frame(name = "Average Purchase Price")

age_avPerson = age_totalValue/(total_age_df['Total Age Count'])
age_avPerson_df = age_avPerson.to_frame(name = "Avg Total Purchase per Person")

age_analysis_df = pd.concat([age_purchaseCount_df,age_app_df,age_totalValue_df,age_avPerson_df], axis = 1)
age_analysis_df
```

```python
#apply formatting
age_analysis_df["Average Purchase Price"] = age_analysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_analysis_df["Total Purchase Value"] = age_analysis_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_analysis_df["Avg Total Purchase per Person"] = age_analysis_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create analysis calculations and convert to dataframes
#found the 2 column groupby here: https://stackoverflow.com/questions/51330281/sum-values-of-column-based-on-the-unique-values-of-another-column 
user_total_value = purchase_data_df.groupby('SN') ['Price'].sum()
user_total_value_df = user_total_value.to_frame(name = "Total Purchase Value")

user_purchase_count = purchase_data_df.groupby('SN') ['Item ID'].count()
user_purchase_count_df = user_purchase_count.to_frame(name = "Purchase Count")

user_app = user_total_value / user_purchase_count
user_app_df = user_app.to_frame(name = "Average Purchase Price")

user_analysis_df = pd.concat([user_purchase_count_df,user_app_df,user_total_value_df], axis = 1)
user_analysis_df.sort_values(by=['Total Purchase Value'],ascending=False).head(5)
```

```python
#find top 5
top_five_df = user_analysis_df.nlargest(5, 'Total Purchase Value')
```

```python
#apply formatting
top_five_df["Average Purchase Price"] = top_five_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_five_df["Total Purchase Value"] = top_five_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
top_five_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create analysis calculations and convert to dataframes
item_count = purchase_data_df.groupby(['Item ID','Item Name']) ['Item ID'].count()
item_count_df = item_count.to_frame(name = "Purchase Count")

item_price = purchase_data_df.groupby(['Item ID','Item Name']) ['Price'].mean()
item_price_df = item_price.to_frame(name = "Price")

total_item_value = purchase_data_df.groupby(['Item ID','Item Name']) ['Price'].sum()
total_item_value_df = total_item_value.to_frame(name = "Total Purchase Value")

#combine onto one DF
item_analysis_df = pd.concat([item_count_df,item_price_df,total_item_value_df], axis = 1)
most_popular_df = item_analysis_df.sort_values(by=['Purchase Count'],ascending=False).head(5)
most_popular_df
```

```python
#apply formatting
most_popular_df["Total Purchase Value"] = most_popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
most_popular_df["Price"] = most_popular_df["Price"].astype(float).map("${:,.2f}".format)
most_popular_df
```
